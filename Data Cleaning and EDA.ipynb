Separate the city and state:

```python
separ = df.location.str.split(',', expand = True)
df['city'] = separ[0]
df['newstate'] = separ[1]

print(df.city)

state1 = df.newstate.str.split(expand = True)
df['state'] = state1[0]
print(df.state)

```

```python
#drop the column location:
df = df.drop(['location'], axis=1)
df = df.drop(['newstate'], axis=1)
```

```python
df.head(5)
```

```python
#20 Companies that have the most reviews
top_review = df.groupby(['company'])['reviews'].sum().reset_index()
top_review.sort_values(['reviews'], ascending = False).head(10)

```

```python
#20 companies that have the most open positions

positions_company = df.groupby(['company'])['position'].count().sort_values(ascending = False).reset_index()
positions_company.head(10)
```

```python

most_positions = pd.DataFrame(df['position'].value_counts())
most_positions

#Another way using of counting them
#df.groupby(['position'])['company'].agg(['count']).sort_values(by='count', ascending=False).reset_index()
```
